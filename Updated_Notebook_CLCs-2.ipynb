```python
#FILTERING THE DATA! In this step we are only remocing the sequences of length less than 300aa and more than 2000aa 

file_path = '/Users/yvettedimitrova/Downloads/unknowns_aftercd-hit.fas.3.clstr.txt'

# Open the file and read its contents
with open(file_path, 'r') as file:
    data = file.read()

lines = data.strip().split('\n')

# Filtering the data
filtered_lines = []
current_cluster = ''
for line in lines:
    if line.startswith('>Cluster'):
        current_cluster = line  # Keep the current cluster line for reference
    else:
        # Extract the length of the protein sequence
        length = int(line.split()[1].rstrip('aa,'))
        if 400 < length < 1200:
            # Keep the protein lines within the specified range along with the cluster info
            filtered_lines.append(current_cluster)  # Add the current cluster line
            filtered_lines.append(line)  # Add the complete protein line

# Joining the filtered lines
filtered_data_updated = '\n'.join(filtered_lines)
print(filtered_data_updated)


downloads_folder_path = '/Users/yvettedimitrova/Downloads/filtered_data_unknowns.txt'  # Update this path
with open(downloads_folder_path, 'w') as output_file:
    output_file.write(filtered_data_updated)


print("Output written to filtered_data_unknowns.txt")
```

```python
#! This chunk of code gives us the final output desired. We are left with one sequence per cluster (only the ones marked with an asterisk *).
#Finally, we are also sorting the sequences left by length in an ascending order, starting with the shorter proteins.

# Splitting the input data into lines
lines = filtered_data_updated.strip().split('\n')

# Tracking clusters, their proteins, and sorting by protein length
cluster_proteins = {}
for line in lines:
    if line.startswith('>Cluster'):
        current_cluster = line
        if current_cluster not in cluster_proteins:
            cluster_proteins[current_cluster] = []
    else:
        length = int(line.split()[1].rstrip('aa,'))
        cluster_proteins[current_cluster].append((length, line))

# Processing and sorting clusters
sorted_data = []
for cluster, proteins in cluster_proteins.items():
    if len(proteins) > 1:
        # Keep only protein with '*' and sort by length
        proteins = sorted([p for p in proteins if p[1].strip().endswith('*')], key=lambda x: x[0])
    else:
        # Sort single entry clusters by length
        proteins = sorted(proteins, key=lambda x: x[0])
    for length, protein in proteins:
        sorted_data.append((length, cluster, protein))

# Sorting all data by protein length
sorted_data = sorted(sorted_data, key=lambda x: x[0])

# Preparing final output
final_output_updated = []
for length, cluster, protein in sorted_data:
    final_output_updated.append(cluster)
    final_output_updated.append(protein)

# Joining the final output
final_output_updated = '\n'.join(final_output_updated)
print(final_output_updated)

downloads_folder_path = '/Users/yvettedimitrova/Downloads/u_final_output_updated.txt'  # Update this path
with open(downloads_folder_path, 'w') as output_file:
    output_file.write(final_output_updated)


print("Output written to u_final_output_updated.txt")
```

```python
#extracting only protein names
import re
protein_names_updated = re.findall(r'>([A-Z0-9]+)\|', final_output_updated)

protein_names_updated

downloads_folder_path = '/Users/yvettedimitrova/Downloads/protein_names_unknowns.txt'
with open(downloads_folder_path, 'w') as output_file:
    # Join the list elements into a single string, each separated by a newline
    output_file.write('\n'.join(protein_names_updated))

print("Output written to protein_names_unknowns.txt")
```
